```python
data2['Salary_USD']=data2['CompConverted']*data2['exchangeRate']

#CompensationAmount
#CompensationCurrency
#exchangeRate
```

```python
data2.head(20)
```

```python
def income(row):
    if row ["Salary_USD"] <=20000:
        return "$20,000 or less"
    elif 20000<(row["Salary_USD"])<=40000:
        return "$20,001-$40,000"
    elif 40000<(row["Salary_USD"])<=60000:
        return "$40,001-$60,000"
    elif 60000<(row["Salary_USD"])<=80000:
        return "$60,001-$80,000"
    elif 80000<(row["Salary_USD"])<=100000:
        return "$80,001-$100,000"
    elif 100000<(row["Salary_USD"])<=150000:
        return "$100,001-$150,000"
    elif 150000<(row["Salary_USD"])<=200000:
        return "$150,001-$200,000"
    elif 200000<(row["Salary_USD"])<=999999999:
        return "Over $200,000"

data2 = data2.assign(income_group=data2.apply(income, axis=1))

data2.head(50)
```

```python
data2.drop(data2[data2.Salary_USD > 300000].index, inplace=True)
```

```python
data2.drop(data2[data2.Salary_USD == 0].index, inplace=True)
```
